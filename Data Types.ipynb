## Normalization

```python
data = np.array([[10, 15, 20], [25, 30, 35], [40, 45, 50], [55, 60, 65], [15, 25, 35]])
```

```python
mean = data.mean(axis=0)
std = data.std(axis=0)
normalized_data = (data - mean) / std
```

```python
normalized_data
```
